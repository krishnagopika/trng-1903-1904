**iterators**

1. iter()
2. next()

```python
my_list = [1,4,5,6]

my_iterator = iter(my_list)

while True:
    try:
        i = next(my_iterator)
        print(i)
    except:
        print("end of list")
        break

for j in range(len(my_list)):
    i = next(my_iterator)
    print(i)
```
